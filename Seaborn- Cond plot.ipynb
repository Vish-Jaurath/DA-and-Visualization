## Conditional Plots using Seaborn

I've worked on Titanic data set to create visualizations using seaborn library to quickly style plots and created small multiples using conditions.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
titanic= pd.read_csv('train.csv')
titanic= titanic[['Survived','Pclass','Sex','Age','SibSp','Parch','Fare','Embarked']]
```

```python
titanic= titanic.dropna()
titanic.head()
```

### Creating Histograms In Seaborn

```python
sns.distplot(titanic['Age'])
```

### Generating A Kernel Density Plot

```python
sns.kdeplot(titanic['Age'], shade= True)
```

### Modifying The Appearance Of The Plots

```python
sns.set_style('white')
sns.kdeplot(titanic['Age'], shade= True)
sns.despine(left= True, bottom= True)
plt.xlabel("Passenger's Age")
```

### Conditional Distributions Using A Single Condition

```python
g= sns.FacetGrid(titanic, col= 'Pclass', height= 6)
g.map(sns.kdeplot, 'Age', shade= True) 
sns.despine(left= True, bottom= True)
```

### Creating Conditional Plots Using Two Conditions

```python
g= sns.FacetGrid(titanic, col= 'Survived', row= 'Pclass', height= 4)
g.map(sns.kdeplot, 'Age', shade= True) 
sns.despine(left= True, bottom= True)
```

### Creating Conditional Plots Using Three Conditions

```python
g= sns.FacetGrid(titanic, col= 'Survived', row= 'Pclass', hue= 'Sex', height= 4)
g.map(sns.kdeplot, 'Age', shade= True) 
sns.despine(left= True, bottom= True)
```

### Adding A Legend

```python
g= sns.FacetGrid(titanic, col= 'Survived', row= 'Pclass', hue= 'Sex', height= 4)
g.map(sns.kdeplot, 'Age', shade= True).add_legend()
sns.despine(left= True, bottom= True)
```
